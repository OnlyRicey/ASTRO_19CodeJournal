Select your favorite distribution (other than uniform and normal) and pull 1000 random numbers from that distribution. Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.




```python
import matplotlib.pyplot as mlib
import numpy as np
```

```python
x = np.random.exponential(scale=1.0, size=1000)
```

```python
mlib.hist(x, bins=100, edgecolor='black')
mlib.xlabel("Number")
mlib.ylabel("Frequency")
mlib.savefig("histogram.pdf", format='pdf')
```
